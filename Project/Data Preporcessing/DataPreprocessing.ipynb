## create interactions dataset


```python
col=t.columns.to_list()
col = col[:2] + col[18:]
interaction = t[col]
interaction
```

```python
col=interaction.columns.to_list()
products = col[2:]
products
```

```python
tmp = pd.melt(interaction, id_vars=["timestamp","user_Id"], value_vars=products, var_name="item_id")
tmp2 = tmp["value"]==1
interaction=tmp[tmp2]
```

```python
interaction= interaction.drop(columns="value")
```

```python
interaction = interaction.drop_duplicates(subset=["user_Id", "item_id"], keep="first")
```

```python
interaction.sort_values(by=["user_Id","timestamp"], inplace=True)
```

```python
interaction["timestamp"]=interaction["timestamp"].apply(lambda x: (x - datetime.datetime(1970, 1, 1)).total_seconds())
```

```python
interaction["user_Id"]=interaction["user_Id"].apply(lambda x: str(x))
```

```python
interaction.dtypes
```

```python
col=interaction.columns.to_list()
col = col[-2:] + col[0:1]
col
interaction=interaction[col]
interaction
```

```python
interaction.to_csv("interaction_v3.csv", index=False)
```
